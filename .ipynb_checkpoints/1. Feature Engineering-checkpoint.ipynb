## Age demographics

Let's try data on age demographics, which can be found here: https://commonslibrary.parliament.uk/home-affairs/communities/demography/constituency-statistics-population-by-age/

```python
df_ages = pd.read_csv('CSVs/Ages.csv')
df_ages.rename(columns={'ONSConstID': "ID",
                       'ConstituencyName':'Name',
                       'RegionName':'Region',
                       'ConstLevel':'Population'}, inplace = True)

print('Dataframe shape:', df_ages.shape)
df_ages.head(3)
```

A few things jump out here:

* Each constituency has its own code, as defined by the Office of National Statistics (ONS). This will help us join tables later on - indeed we should consider using this as the index for our constituency dataframe.
* The dataframe is in long format, which we do not want. We will need to work out how to deal with this.
* The dataframe comes with lots of columns that we don't need, for example, metrics calculated at a regional and national level.

As a starting point, let's define a master dataframe, which will house all the information we need from the various CSV files.

```python
#Define columns to start with
cols = ['ID','Name','Region']

#Drop duplicates caused by multiple years/age groups in the age data
df_basics = df_ages[cols].drop_duplicates()
df_basics.set_index('ID', drop=True, inplace=True)
df_basics.head(3)
```

Getting back to the ages dataframe, let's just keep the data that we need.

```python
#Define the columns we want to keep for now
cols = ['ID', 'Age group', 'Population', 'Const%', 'Date']
df_ages = df_ages[cols]
print('Dataframe shape:', df_ages.shape)
df_ages.head(3)
```

We clearly have more rows than we should. Of course, we have 650 constituencies, but we also have multiple age groups, and we also seem to have multiple dates for this data.

```python
print('Unique Age Ranges:', df_ages['Age group'].unique())
print('Unique Years:', df_ages['Date'].unique())
```

However, we see that shares within age ranges tend to not move around much in the space of only six years.

```python
df_ages.loc[(df_ages['ID']=='E14000569') &
            (df_ages['Age group'] == '20-29')].sort_values('Date')
```

Thus, we can probably get rid of historical data for this KPI without losing much information.

```python
df_ages = df_ages.loc[df_ages['Date']==2018]
print(df_ages.shape)
```

One key use for this data is to infer the total population of each constituency. This will be useful later if we need to relitavise any metrics on a per capita basis.

```python
#Sum the age groups by constituency
population = df_ages.groupby('ID').sum()[['Population']]

#Join the populations to the master dataframe
df = pd.concat([df_basics,population], axis=1)
df.head(3)
```

We note that the UK is broken down into ten or so regions. Let's see how these compare by population.

```python
pop_regions = df.groupby('Region')[['Population']].sum().sort_values('Population')

pop_regions.plot.barh(figsize=(15,5), width=0.8, color=CB91_Amber)

plt.title('Total Population by Region (2019)')
plt.ylabel('')
plt.legend().remove();
```

Going back to our age groups, we now have the task of getting the dataframe from a long to a wide format, so that the age ranges are turned into columns. To do this, we can use a pivot table.

```python
#Define the pivot table
ages_pivot = pd.pivot(df_ages, index='ID',
                      columns='Age group',
                      values='Const%')

#Turn the percentage strings into floats
ages_pivot = ages_pivot.applymap(lambda x: float(x[:-1]) / 100)

#Rename the columns
ages_pivot.columns = [f'%{i}YO' for i in ages_pivot.columns]
```

We can now join this table onto our master dataframe.

```python
df = pd.concat([df,ages_pivot], axis=1)
df.head(3)
```

Let's see what the UK's age distribution looks like on a region by region basis.

```python
#Group by region to find the mean share of each age category
stacked_ages = df.groupby('Region').mean().sort_values('%0-9YO').iloc[:,1:]

#Plot the data
stacked_ages.plot.barh(stacked=True, figsize=(15,5), width=0.8);

plt.xticks(ticks = np.arange(0,1.2,0.2), labels = np.arange(0,120,20))
plt.xlim(0,1);

plt.title('Age Distribution by Region');
plt.ylabel('');

plt.legend(bbox_to_anchor=(0.95, -0.1),
           ncol=9,
           frameon=False,
           labels=['% 0-9','% 10-19','% 20-29',
                   '% 30-39','% 40-49','% 50-59',
                   '% 60-69','% 70-79','% 80+']);
```

### Mapping Data

It would be nice to have some kind of map visual showing us all the data that we are gathering. We could take a standard approach to mapping this, using the constituencies exact geographies, though we've already seen that the vastly different areas could make such maps misleading.

For example, if we took England in isolation and mapped which party had won in each constituency in 2017, it might appear that The Conservatives (in blue) absolutely monstered the election.

<img src="images/2017_England_Election.png" alt="Election Map" style="width: 600px;"/>


In fact, only 41% of those constituencies are shaded blue! If we zoom in on the cities, we see that Labour (in red) won in most of the geographically 'small' constituencies. Remember - they all count the same in Parliament.

We therefore need a map that makes each constituency the same size. These are commonly used in broadcast visualisations (the BBC has a giant real-life version outside its HQ on election night!)

![title](images/BBC_Hex.png)

To start doing this, we can modify some existing JSON code, which was sourced here: https://github.com/odileeds/hexmaps/blob/gh-pages/maps/constituencies.hexjson

This JSON file assigns a 'co-ordinate pair' to each constituency on a 45x30 grid. We should be able to use these co-ordinates to plot a scatter chart that looks like a map of the UK, with each constitutnecy represented by one marker on a co-ordinate.

When generalising this to a function that will map KPIs, we will also want the map to colour-code itself according to the KPI that we pass.

Let's create a series of functions to do this.

```python
def gradient_mapper(kpi, grad, outliers=None, stretch=1, factor=1):
    
    '''
    Takes a list/series of numbers, outputs a list of hex colours,
    appropriate for heatmapping the initial data.
    
    Parameters:
    - col (list/series, etc.): The data to be transformed
    - grad (list hex codes): A list of colors that the data will be transformed to
    - outliers (top,bottom,both): Stretches the outliers, resulting in more gradient
                                  change amongst clustered values
    - stretch (int): The number of colors to duplicate if outliers variable used
    - factor (int): The scale of color duplication if outliers variable used
    
    '''
    #Work out how many colours we have in the given gradient
    colors = len(grad)
    half = colors // 2
    
    #Ensure that stretch is possible
    stretch = min(half//3, stretch)
    
    factors = [4*factor, 3*factor, 2*factor]
    
    if outliers != None:
        #Stretch gradient if required. Declare three lists:
        #Start is the stretch map for the bottom end
        if (outliers.lower() == 'bottom') or (outliers.lower() == 'both'):
            start = [factors[0]]*stretch + [factors[1]]*stretch + [factors[2]]*stretch
        else:
            start = []

        #End is the stretch map for the top end
        if (outliers.lower() == 'top') or (outliers.lower() == 'both'):
            end = [factors[2]]*stretch + [factors[1]]*stretch + [factors[0]]*stretch
        else:
            end = []

        #Middle is a list of 1s which will be non-transformed
        middle = [1 for i in range(colors - len(start) - len(end))]

        stretch_map = start + middle + end
        
    else:
        stretch_map = [1 for i in range(colors)]
        
    #Create tuples of the gradients, and the number of
    #times they should be repeated in the list
    zip_list = list(zip(grad,stretch_map))
    
    #Use this to create a list of lists
    #Each element will be a list of the same gradient
    #repeated the required number of times
    list_of_lists = [[i[0]]*i[1] for i in zip_list]
    
    #Melt this list of lists into a single list
    grad = sum(list_of_lists, [])
    
    #Re-define colors variable
    colors = len(grad)-1
    
    #Define the lowest and the highest points in the dataset
    kpi_min = kpi.min()
    kpi_max = kpi.max()
    
    #Transform the data to integers between zero and the length of the gradient list
    first_map = list(map(lambda x: int(round(colors*(x-kpi_min) /
                                             (kpi_max-kpi_min),0)), list(kpi)))
    
    #Map the integers onto the gradient list
    second_map = list(map(lambda x: grad[x], first_map))
    
    #Return this, as well as the new gradient
    return second_map, grad




f = open('Datasets/constituencies.hexjson')
datamap = json.load(f)
datamap = pd.DataFrame(datamap['hexes']).T
datamap = datamap[['n','q','r']]
datamap.columns=['Name','X','Y']

def kpi_map(kpi, width=6, colorbar=True, outliers=None, stretch=1, factor=1,
            exclude=[], title=None, colors=None, grad=CB91_Grad_AP, data=df):
    
    '''
    Outputs a choropleth map, showing each constituency in the same size.
    
    Parameters:
    - kpi (list/series, etc.): The data to be transformed
    - width (float): The desired width of the figure
    - df (dataframe): The dataframe to get the data from
    - grad (list hex codes): A list of colors that the data will be transformed to
    - outliers (top,bottom,both): Stretches the outliers, resulting in more gradient
                                  change amongst clustered values
    - stretch (int): The number of colors to duplicate if outliers variable used
    - factor (int): The scale of color duplication if outliers variable used
    - exclude (list): A list of regions to exclude from the chart
    - colors (list): A list of hex-codes to pass
    - title (string): The desired title of the chart
    
    '''    
    
    #Filter out different regions, depending on paramaters
    df_filtered = pd.concat([datamap, data[['Region',kpi]]], axis=1)
    df_filtered = df_filtered.loc[~data['Region'].isin(exclude)]
    
    kpi_filtered = df_filtered[kpi]
    
    if colors == None:
        #Use the gradient mapper function to return the colors for the plot
        gradient_map = gradient_mapper(kpi=kpi_filtered,
                                    grad=grad,
                                    outliers=outliers,
                                    stretch=stretch,
                                    factor=factor)
        colors = gradient_map[0]
    
    #Work out the aspect ratio of the filtered constituencies
    X_diff = np.max(df_filtered['X'])-np.min(df_filtered['X'])
    Y_diff = np.max(df_filtered['Y'])-np.min(df_filtered['Y'])
    
    #Declare the width and height of the plot
    height = width * (Y_diff/X_diff)
    size = 500*math.pi*((width/X_diff)**2)
    
    #Create the figure
    fig, ax = plt.subplots(figsize=(width,height))
    
    plt.xticks([])
    plt.yticks([])

    #Plot the scatter
    ax1 = fig.add_subplot(1,1,1)
    ax1.scatter(df_filtered['X'],
                df_filtered['Y'],
                s=size,
                marker='s',
                c=colors)
    
    #Remove axes
    sns.despine(left=True, bottom=True)
    ax1.set_title(title);
    
    #plot the colorbar
    if colorbar == True:        
        cmap = LinearSegmentedColormap.from_list(name= '',
                                                 colors=gradient_map[1])
        ax2 = fig.add_subplot(2,30,28)
        norm = mpl.colors.Normalize(vmin=df_filtered[kpi].min(),
                                    vmax=df_filtered[kpi].max())
        cb = mpl.colorbar.ColorbarBase(ax2, cmap=cmap,
                                       norm=norm, orientation='vertical')
    
    # remove the x and y ticks
    for ax in [ax1,ax2]:
        ax.set_xticks([])
        ax.set_yticks([])
```

Let's plot the population density metric to see how this function could work.

```python
kpi_map('PopDensity', outliers='top',
        stretch=3, factor=1,
        title='Population Density by Constituency\nPeople per km^2')
```

So we see that population density in London is very high (we shouldn't be surprised). We also see urban centres such as Bristol, Birmingham, and Liverpool show up in purple.

```python
df_business = pd.read_csv('CSVs/Businesses.csv')
df_business.rename(columns={'ONSConstID': "ID",
                       'ConstituencyName':'Name',
                       'RegionName':'Region',
                       'BusConstNumber':'Businesses',
                       'DateOfDataset':'Date'}, inplace = True)
print(df_business.shape)
df_business.head(3)
```

We have 6,500 rows, which suggests that we have 10 rows per constituency. It is likely that this is the 'Date of Dataset' column, but let's find out for sure.

```python
df_business['Date'].unique()
```

Let's interrogate this data a bit further. Firstly, let's have a look at the spread of businesses across constituencies.

```python
business_2019 = df_business.loc[df_business['Date']=='15/03/2019']

plt.figure(figsize=(12,4))
plt.hist(business_2019['Businesses'], bins=150)
plt.xlim(1000,10000)

plt.title('Histogram of Number of Businesses in Each Constituency')
plt.xlabel('Number of Businesses')
plt.ylabel('Number of Constituencies');
```

So most constituencies have 2,500-4000 businesses.

Given our earlier note about London, let's see how this breaks down at a regional level.

```python
business_regions = df_business.groupby('Region').median().sort_values('Businesses')

business_regions.plot.barh(figsize=(6,10), width=0.8, color=CB91_Amber)

plt.title('Median Number of Incorporated Businesses\nin Constituency by Region (2018)')
plt.ylabel('')
plt.legend().remove();
```

So we see that London constituencies have a high average number of businesses, as we would expect.

What about the evolution over time?

```python
#Get time data into wide format
business_pivot = pd.pivot(df_business, index='ID',
                          columns='Date',
                          values='Businesses')

#Group by region, so we have the regional view over time
business_ts = pd.concat([df_basics,business_pivot],
                        axis=1).groupby('Region').sum()

business_ts.sort_values(business_ts.columns[-1],
                        ascending=False,
                        inplace=True)
```

```python
business_ts.T.plot(figsize=(8,10));
plt.title('Total Number of Incorporated Businesses by Region (2018)');
plt.ylabel('')
plt.legend(bbox_to_anchor=(1.1, -0.1),
           ncol=4,
           frameon=False);
```

We see that the total number of businesses has grown significantly in London, and quite strongly in the South East. This certainly plays into the idea that economic gains since the financial crisis have not been evently spread.

### Scotland & Northern Ireland Data

This is frustrating. Indeed, it seems that house price data at a Scottish / Northern Irish constituency level is difficult to come by. However, we are able to find it at a 'Local Authority' level. Note - this is a 'layer above' constituency level. We can therefore get a feel for house prices in the area.

```python
df_LAhouse = pd.read_csv('CSVs/LocalAuthorityHousePrices.csv')
df_LAhouse.rename(columns={'Area_Code': "LA Code",
                       'ConstituencyName':'Name',
                       'Region_Name':'Region',
                       'Average_Price':'HousePrice'}, inplace = True)

#Set the date column as a datetime object
df_LAhouse['Date'] = pd.to_datetime(df_LAhouse['Date'])

#Keep the columns we need
df_LAhouse = df_LAhouse[['LA Code','Region','Date','HousePrice']]

#We can immediately get rid of everything from before 1995,
#and cases where it's just aggregate data for Scotland / NI
df_LAhouse = df_LAhouse.loc[(df_LAhouse['Date'] > '1995')
                           & (df_LAhouse['Region'] != 'Scotland')
                           & (df_LAhouse['Region'] != 'Northern Ireland')]

#And we only need to keep Scotland and Northern Ireland data
df_LAhouse = df_LAhouse.loc[(df_LAhouse['LA Code'].str.startswith('N'))
                          | (df_LAhouse['LA Code'].str.startswith('S'))]

df_LAhouse.head()
```

We now need to bring in a table that shows us how Local Authority IDs map onto the constituency ones.

```python
code_map = pd.read_csv('CSVs/CodeLookup.csv')
code_map.head()
```

Let's create a function that uses this table to take a constituency code, and return its local authority code.

Then a second function that will return a local authority price for a given date.

```python
def LALookUp(ConID):
    return code_map.loc[code_map['ID'] == ConID]['LA Code'].item()

def LAPriceLookUp(ConID, date):
    LA_Code = LALookUp(ConID=ConID)
    try:
        return df_LAhouse.loc[(df_LAhouse['LA Code'] == LA_Code)
              &(df_LAhouse['Date'] == date)]['HousePrice'].item()
    except:
        return None
```

We can use these to back-fill some of the missing data from the original house price dataframe.

```python
#Identify the Scottish and Northern Irish constituency IDs
NISIDs = df.loc[(df['Region']=='Scotland') |
                (df['Region']=='Northern Ireland')].index

#For each seat for which we're still missing data...
for i in NISIDs:
    #For each date...
    for j in df_house['Date'].unique():
        #Update price with local authority data
        df_house.loc[(df_house['ID'] == i) &
                     (df_house['Date'] == j), 'HousePrice'] = LAPriceLookUp(i, j)
```

Let's create a pivot table to view the time series data in a wide format.

```python
house_pivot = pd.pivot(df_house, index='ID',
                          columns='Date',
                          values='HousePrice')

for i in house_pivot.columns:
    house_pivot[i] = pd.to_numeric(house_pivot[i])
    
house_ts = pd.concat([df_basics,house_pivot], axis =1)

house_ts = house_ts.groupby(
    'Region').mean().sort_values(house_pivot.columns[-1], ascending=False)

house_ts.T.plot(figsize=(10,9));
plt.title('Average Median Constituency House Price by Region');
plt.ylabel('')
plt.ylim(0,)
plt.legend(bbox_to_anchor=(0.95, -0.1),
           ncol=4,
           frameon=False);
```

There are a couple of very clear takeaways from this chart. The first is that London's property prices have grown much quicker than anywhere else in the country, with the average median house price in London constituencies costing over half a million pounds.

The second thing is the boom and bust of Northern Ireland's property market. This coincided with a similar rise and fall in the Republic of Ireland, though it's notable that prices now are basically as they were in 2005.

```python
#Create a column that looks at house price percentage growth since 2010
house_pivot['Growth'] = house_pivot['2018-01-01 00:00:00'] / house_pivot['2010-01-01 00:00:00'] - 1

#Isolate the last two columns and rename them
house_pivot = house_pivot.iloc[:,-2:]
house_pivot.columns = ['HousePrice','%HousePriceGrowth']

#Add them to the main dataframe
df = pd.concat([df, house_pivot], axis=1)

df.info()
```

### Housing Tenure

Of course, the cost of housing can seriously affect whether or not people can buy. We should therefore expect some disparity in housing tenure and home ownership.

We can get a view of this from the 2011 census. This is not the most up-to-date source, but should add helpful context to the dataset nonetheless.

Data is sourced from here: https://commonslibrary.parliament.uk/social-policy/housing/home-ownership/constituency-data-housing-tenure/

The following defintions are used for tenure:
* Owners may own their home outright or with a mortgage/loan.
* Private renters either rent from a landlord/agency or another party; other private rented refers to arrangements such as renting from an employer.
* Social renters may rent from the local authority or from another social housing provider (typically a housing association).
* Living rent free refers to households living in property owned by another party without paying rent.
* Shared ownership is a type of housing tenure where the household part-owns the property and pays rent on the rest.

```python
df_tenure = pd.read_csv('CSVs/HousingTenure.csv')
df_tenure.rename(columns={'ONSConstID': "ID",
                         'ConstituencyName':'Name',
                         'RegionName':'Region',
                         'CON%Own_out':'%OwnOutright',
                         'CON%Own_mort':'%OwnWithMort',
                         'CON%Share':'%SharedOwn',
                         'CON%Soc_r':'%SocialHousing',
                         'CON%Private_rent':'%PrivateRent',
                         'CON%Rent_free':'%RentFree'}, inplace = True)

#Select the columns we need
df_tenure = df_tenure[['ID','Name','Region','%OwnOutright',
               '%OwnWithMort','%PrivateRent','%SocialHousing',
               '%SharedOwn','%RentFree']]

df_tenure.set_index('ID', drop=True, inplace=True)

df_tenure.head(3)
```

We have some scope to combine these into larger categories. Let's define a new list of industries, which we'll put into some kind of 'blue collar to white collar' order.

```python
industries = ['%Heavy Industry & Manufacturing',
              '%Wholesale & Retail',
              '%Agriculture',
              '%Hospitality',
              '%Public Admin & Defence',
              '%Transport',
              '%Health & Social Work',
              '%Support & Admin',
              '%Education',
              '%Sciences',
              '%FS & ICT',
              '%Other Industry']

industry_key = {' Industry [Census 2011]|A Agriculture; forestry and fishing (percent)':
 '%Agriculture',
 ' B Mining and quarrying (percent)':'%Heavy Industry & Manufacturing',
 'C Manufacturing (percent)':'%Heavy Industry & Manufacturing',
 'D Electricity; gas; steam and air conditioning supply (percent)':
 '%Heavy Industry & Manufacturing',
 ' E Water supply; sewerage; waste management and remediation activities (percent)':
 '%Heavy Industry & Manufacturing',
 'F Construction (percent)':'%Heavy Industry & Manufacturing',
 ' G Wholesale and retail trade; repair of motor vehicles and motor cycles (percent)':
 '%Wholesale & Retail',
 'H Transport and storage (percent)':'%Transport',
 'I Accommodation and food service activities (percent)':'%Hospitality',
 'J Information and communication (percent)':'%FS & ICT',
 'K Financial and insurance activities (percent)':'%FS & ICT',
 'L Real estate activities (percent)':'%Other Industry',
 'M Professional; scientific and technical activities (percent)':'%Sciences',
 'N Administrative and support service activities (percent)':'%Support & Admin',
 'O Public administration and defence; compulsory social security (percent)':'%Public Admin & Defence',
 'P Education (percent)':'%Education',
 'Q Human health and social work activities (percent)':'%Health & Social Work',
 'R - U Other (percent)':'%Other Industry'}


#We create new columns for each of the new headings that we defined
#using the values in the above dictionary.

#Iterate through these values...
for i in industries:
    #Create a new empty column
    df_indus[i] = np.zeros(650)
    
    #Then iterate through each existing column...
    for j in industry_key.keys():
        #And if we want to add it to the new columns, then do so
        if industry_key[j] == i:
            df_indus[i] += df_indus[j]
            
#We re-declare the dataframe, removing the old columns

df_indus = pd.concat([df_basics,df_indus.iloc[:,-12:]], axis=1)
df_indus.head(3)
```

Broadband connections are an essential part of a modern economy. However, rollout in the UK has not been equal, and speeds can vary quite drastically.
```python
df_bb = pd.read_csv('CSVs/Broadband.csv')

df_bb.rename(columns={'pcon-code':'ID',
                     'Speed':'MbpsSpeed',
                     'Under10':'%Under10Mbps',
                     'UnableDecent':'%UnableDecent'}, inplace=True)
df_bb.drop('pcon-nm', inplace=True, axis=1)
df_bb.set_index('ID', inplace=True, drop=True)

print(df_bb.shape)
df_bb.head()
```

### Local Authority Spending

Financial austerity has been a key feature of UK politics in the last decade. It has resulted in the cutting of public services, and is frequently cited as a cause of the Brexit vote. We can get a sense of this by looking at the spending power of local authorities.

This data is available direct from the government here: https://commonslibrary.parliament.uk/home-affairs/communities/local-authority-data-finances/

```python
df_local = pd.read_csv('CSVs/England/LocalGovernmentFinance.csv')
df_local.dropna(subset=['LocalAuthority'])
print(df_local.shape)
df_local.head(3)
```

```python
#We have some naming inconsistencies here,
#so we need to rename some of the authorities

def la_rename(X):
    
    key = {'Bath & North East Somerset':'Bath and North East Somerset',
         'Brighton & Hove':'Brighton and Hove',
         'Bristol':'Bristol, City of',
         'Cheshire West & Chester':'Cheshire West and Chester',
         'Durham':'County Durham',
         'Herefordshire':'Herefordshire, County of',
         'Isle of Wight Council':'Isle of Wight',
         'Kingston upon Hull':'Kingston upon Hull, City of',
         'Shepway':'Folkestone and Hythe',
         'St Helens':'St. Helens',
         'Telford and the Wrekin':'Telford and Wrekin'}
    
    if X in list(key.keys()):
        return key[X]
    
    else:
        return X
    
#Apply the above function
df_local['LocalAuthority'] = [la_rename(i) for i in df_local['LocalAuthority']]

#Set the index
df_local.set_index('LocalAuthority', inplace=True, drop=True)

#Pull the LA Codes from our existing key
LA_Code = []
for i in df_local.index:
    try:
        new_code = code_map.loc[code_map['Local Authority Name']==i].iloc[0,:]['LA Code']
        LA_Code.append(new_code)
    except:
        LA_Code.append('Unknown')
        
df_local['LA Code'] = LA_Code
```

```python
#Drop rows we don't need
to_drop = df_local.loc[df_local['LA Code']=='Unknown'].index
df_local.drop(labels=to_drop, inplace=True)

#Select the columns that we want to run with
df_local = df_local[['% Change 2015/16 to 2018/19', 'LA Code']]

df_local.set_index('LA Code', inplace=True, drop=True)

df_local.head()
```

Let's visualise the growth in Local Authority spending.

```python
plt.figure(figsize=(10,3))
plt.hist(df_local['% Change 2015/16 to 2018/19']*100, bins=40, density=True)
plt.vlines(x=0, ymin=0, ymax=0.11, color='red')

plt.title('% Growth in Local Authority Spending Power (2015-2019)')
plt.xlabel('Spending Power Growth (%)')
plt.ylabel('Share of Local Authorities');
```

So we see that, indeed, most local authorities in the country have seen a real terms reduction in spending power.

```python
(df_health*100).mean().sort_values().plot.barh(figsize=(12,8),
                                         width=0.8);
plt.title('Incidences of Chronic Illnesses (Constituency Means, 2017)')
plt.xlabel('Share of Constituents With Illness');
```

```python
df_health['DiseasesPerPop'] = df_health.sum(axis=1)

df_health = df_health[['High Blood Pressure',
                      'Depression',
                      'Obesity',
                      'Diabetes',
                      'Asthma',
                      'DiseasesPerPop']]
```

We add this to our England dataframe.

```python
df_eng = pd.concat([df_eng, df_health], axis=1)
```
